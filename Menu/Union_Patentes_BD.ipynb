```python
# Librerias
import pandas as pd
import unidecode
```

```python
# Importamos los archivos
df_patente67 = pd.read_csv('INVENTOR6724.csv')
df_patenteMX = pd.read_excel('INVENTORES_MX.xlsx')
```

Hacemos modificaciones en 'df_patente67' tomamos 'inventor_first_name' y 'inventor_last_name' pra unirlas en una sola columna llamada 'NOMBRE_INVENTOR' en el siguiente formato 'inventor_last_name','inventor_first_name'

```python
# Creamos 'NOMBRE_INVENTOR'
df_patente67['NOMBRE_INVENTOR'] = df_patente67['inventor_last_name'] + ',' + df_patente67['inventor_first_name']
# Creamos 'Inventor'
df_patente67['Inventor'] = df_patente67['inventor_first_name'] + ' ' + df_patente67['inventor_last_name']

# Eliminamos guiones, comas, numero y puntos en la columna 'NOMBRE_INVENTOR' y 'Inventor'
df_patente67['NOMBRE_INVENTOR'] = df_patente67['NOMBRE_INVENTOR'].str.replace('-', ' ').str.replace('[0-9]', '').str.replace('.', '').apply(unidecode.unidecode)
df_patente67['Inventor'] = df_patente67['Inventor'].str.replace('-', ' ').str.replace('[0-9]', '').str.replace('.', '').apply(unidecode.unidecode)

# Eliminamos guiones, comas, numero y puntos en la columna 'NOMBRE_INVENTOR' y 'Inventor'
df_patenteMX['NOMBRE_INVENTOR'] = df_patenteMX['NOMBRE_INVENTOR'].str.replace('-', ' ').str.replace('[0-9]', '').str.replace('.', '').apply(unidecode.unidecode)
df_patenteMX['Inventor'] = df_patenteMX['Inventor'].str.replace('-', ' ').str.replace('[0-9]', '').str.replace('.', '').apply(unidecode.unidecode)

# Eliminamos las columnas que no necesitamos
df_patente67 = df_patente67.drop(columns=['inventor_first_name', 'inventor_last_name'])
```

Hacemos modificaciones en 'df_patente67' tomamos 'inventor_city' y 'inventor_country' para unirla en una sola columna llamada 'Location' en el siguiente formato 'inventor_city','inventor_country'

```python
# Creamos 'Location'
df_patente67['Location'] = df_patente67['inventor_city'] + ', ' + df_patente67['inventor_country']

# Eliminamos las columnas que no necesitamos
df_patente67 = df_patente67.drop(columns=['inventor_lastknown_city', 'inventor_lastknown_country','inventor_country', 'inventor_city'])
```

```python
# Cambiamos el nombre de las columnas en 'df_patente67'
df_patente67.rename(columns={'inventor_total_num_patents': 'Patents', 'patent_number':'numero patente','patent_title':'titulo patente','patent_date':'fecha patente'}, inplace=True)
```

Sacamos un posible años de actividad en 'df_patente67' agarrando solamente el año de la columna 'inventor_first_seen_date' y el de 'inventor_last_seen_date' en el siguiente formato '2019-2020'

```python
# Creamos 'years active'
df_patente67['years active'] = df_patente67['inventor_first_seen_date'].str[:4] + '-' + df_patente67['inventor_last_seen_date'].str[:4]

# despues de ahi sacamos las siguientes columnas 'año de inicio' y 'última patente' separando los años de 'Years active'
df_patente67['año de inicio'] = df_patente67['years active'].str[:4].astype(int)
df_patente67['última patente'] = df_patente67['years active'].str[5:].astype(int)
```

```python
# Eliminamos las columnas que no necesitamos en 'df_patente67'
df_patente67 = df_patente67.drop(columns=['inventor_id', 'inventor_first_seen_date', 'inventor_last_seen_date' , 'inventor_sequence', 'inventor_state', 'inventor_state_fips'])

# Eliminamos las columnas que no necesitamos en 'df_patenteMX'
df_patenteMX = df_patenteMX.drop(columns=['LinkedIn', 'Correo electrónico', 'Estado civil', 'Institución agrupado '])
```

Vemos que la columna 'Patents' es obj, pero la necesitamos en numero para hacer un merge correcto. 

Vemos estos valores
- 3 act 4
- 3 de 4
- 3 act 1

para hacer un mejor merge, se les dara el valor de '3'

```python
table = df_patenteMX['Patents'].value_counts()
table
```

```python
# convertimos los valores de la columna 'patents' que contengan '3 act 4', '3 de 4' y '3 act 1' a 3
df_patenteMX['Patents'] = df_patenteMX['Patents'].replace(['3 act 4', '3 de 4', '3 act 1'], 3).astype(int)
```
